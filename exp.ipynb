```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_players = pd.read_csv("scraping/players.csv")
df_salaries = pd.read_csv("scraping/salaries.csv")
```

```python
df = df_salaries.merge(df_players, left_on="link", right_on="link")
df["normalized_salary"] = df["salary"] / df["salary_cap"]

def get_season_number(x):
    df_link = df[df["link"] == x["link"]]
    return x["season"] - min(df_link["season"])

df["season_number"] = df.apply(get_season_number, axis=1)
```

```python
df = df[df["position"] != "Unknown"]

ranges = [(1987,1995), (1992,2000), (1997,2005), (2003,2010), (2008,2013), (2014, 2020)]
for r in ranges:
    df_year = df[df["year"] >= r[0]]
    df_year = df_year[df_year["year"] < r[1]]

    d = {
        "Season Number": [], 
        "# Center": [], 
        "# Other": [],
        "Center Mean": [], 
        "Other Mean": [],
        # "Center 50": [], 
        # "Other 50": [],
        "Center 75": [], 
        "Other 75": [],
    }
    for num in sorted(df_year["season_number"].unique()):
        df_num = df_year[df_year["season_number"] == num]
        df_center = df_num[df_num["position"].str.contains("Center")]
        df_other = df_num[~df_num["position"].str.contains("Center")]
        d["Season Number"].append(num)
        d["# Center"].append(df_center["normalized_salary"].count())
        d["# Other"].append(df_other["normalized_salary"].count())
        d["Center Mean"].append((df_center["normalized_salary"] * 100).mean())
        d["Other Mean"].append((df_other["normalized_salary"] * 100).mean())
        # d["Center 50"].append((df_center[(df_center["normalized_salary"] >= df_center["normalized_salary"].quantile(0.25)) & (df_center["normalized_salary"] <= df_center["normalized_salary"].quantile(0.75))]["normalized_salary"] * 100).mean())
        # d["Other 50"].append((df_other[(df_other["normalized_salary"] >= df_other["normalized_salary"].quantile(0.25) & (df_other["normalized_salary"] <= df_other["normalized_salary"].quantile(0.75)))]["normalized_salary"] * 100).mean())
        d["Center 75"].append((df_center[df_center["normalized_salary"] >= df_center["normalized_salary"].quantile(0.75)]["normalized_salary"] * 100).mean())
        d["Other 75"].append((df_other[df_other["normalized_salary"] >= df_other["normalized_salary"].quantile(0.75)]["normalized_salary"] * 100).mean())
    print(r)
    display(pd.DataFrame(d))
    df_plt = pd.DataFrame(d)
    plt.figure()
    plt.plot(df_plt["Season Number"], df_plt["Center Mean"])
    plt.plot(df_plt["Season Number"], df_plt["Other Mean"])
    plt.show()
    plt.figure()
    plt.plot(df_plt["Season Number"], df_plt["Center 75"])
    plt.plot(df_plt["Season Number"], df_plt["Other 75"])
    plt.show()
```
